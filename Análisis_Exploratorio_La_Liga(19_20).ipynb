```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
newdf=pd.read_csv("transfermarkt_fbref_201920.csv",sep=';')
```

```python
la_liga_data = newdf[newdf['league'] == 'La Liga']
```

### 3.3 Análisis exploratorio

#### 3.3.1 Penalties

```python
#Penalty goals
Total_Penaltygoals=la_liga_data['pens_made'].sum()
Total_Penaltyatt=la_liga_data['pens_att'].sum()
print(Total_Penaltygoals)
print(Total_Penaltyatt)
```

```python
import seaborn as sns
#Pie chart for penalties missed vs scored
plt.figure(figsize=(13,6))
pl_not_scored=Total_Penaltyatt-Total_Penaltygoals
data=[pl_not_scored,Total_Penaltygoals]
labels=['Penalties missed','Penalties scored']
color=sns.color_palette('Set2')
plt.pie(data,labels=labels,colors=color,autopct='%.0f%%')
plt.show()
```

#### 3.3.2 Posiciones

```python
# Unique positions
la_liga_data['position'].unique()
```

```python
# Total DF players
la_liga_data[la_liga_data['position']=='DF']
```

#### 3.3.3 Nacionalidades

```python
# Player for different nations

np.size((la_liga_data['nationality'].unique()))
```

```python
# Most players from which countries

nationality=la_liga_data.groupby('nationality').size().sort_values(ascending=False)
nationality.head(10).plot(kind='bar',figsize=(12,6), color=sns.color_palette("magma"))
```

#### 3.3.4 Edad

```python
# Players based on age group

Sub21=la_liga_data[la_liga_data['age']<=21]
Age21_27=la_liga_data[(la_liga_data['age']>21)&(la_liga_data['age']<=27)]
Above_27=la_liga_data[la_liga_data['age']>27]
```

```python
x=np.array([Sub21['player'].count(),Age20_27['player'].count(),Above_27['player'].count()])
mylabels=["<=21","Between 21-27",">27"]
plt.title('Total Players with Age Groups', fontsize=20)
plt.pie(x,labels=mylabels,autopct="%.1f%%")
plt.show()
```

```python
# Total players Under 21 in each club
players_under_21 =la_liga_data[la_liga_data['age']<=21]
players_under_21['squad'].value_counts().plot(kind='bar', color=sns.color_palette("cubehelix"))
```

```python
# Players under 21 FC Barcelona
players_under_21[players_under_21["squad"]=="Barcelona"]
```

```python
# Players under 21 Atlético
players_under_21[players_under_21["squad"]=="AtlĂ©tico Madrid"]
```

```python
# Players under 21 Real Madrid
players_under_21[players_under_21["squad"]=="Real Madrid"]
```

```python
# Average age of players in each club

plt.figure(figsize=(12,6))
sns.boxplot(x='squad',y='age', data=la_liga_data)
plt.xticks(rotation=90)
```

```python
num_player=la_liga_data.groupby('squad').size()
data=(la_liga_data.groupby('squad')['age'].sum())/num_player
data.sort_values(ascending=False)
```

#### 3.3.5 Asistencias

```python
# Total assists from each club

Assist_by_clubs=pd.DataFrame(la_liga_data.groupby('squad',as_index=False)['assists'].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax=sns.barplot(x='squad',y='assists',data=Assist_by_clubs.sort_values(by="assists"),palette='Set2')
ax.set_xlabel("squad",fontsize=30)
ax.set_ylabel("assists",fontsize=30)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Assists',fontsize=20)
```

```python
# Top 10 Assists

top_10_assists=la_liga_data[['player','squad','assists','games']].nlargest(n=10,columns='assists')
top_10_assists
```

#### 3.3.6 Goles

```python
Goals_by_clubs=pd.DataFrame(la_liga_data.groupby('squad',as_index=False)['goals'].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax=sns.barplot(x='squad',y='goals', data=Goals_by_clubs.sort_values(by="goals"),palette='rocket')
ax.set_xlabel("squad", fontsize=30)
ax.set_ylabel("goals", fontsize=30)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Goals',fontsize=20)
```

```python
# Most goals by players
top_10_goals=la_liga_data[['player','squad','goals','games']].nlargest(n=10,columns='goals')
top_10_goals
```

```python
# Pie chart - Goals with assist and without assist

plt.figure(figsize=(14,7))
Total_goals=la_liga_data['goals'].sum()
assists=la_liga_data['assists'].sum()
data=[Total_goals-assists, assists]
labels=['Goals witout assists','Goals with assists']
color=sns.color_palette('Set1')
plt.pie(data,labels=labels,colors=color, autopct='%.0f%%')
plt.show()
```

#### 3.3.7 Tarjetas amarillas

```python
# Top 10 players with most yellow cards

liga_yellow=la_liga_data.sort_values(by='cards_yellow',ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title("Players with the most yellows cards")
c=sns.barplot(x=liga_yellow['player'],y=liga_yellow['cards_yellow'],label='Players',color='yellow')
plt.ylabel('Number of Yellow cards')
c.set_xticklabels(c.get_xticklabels(),rotation=45)
c
```
